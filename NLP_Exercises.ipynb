```python
def extract_ne(quote, language = 'english'):
    words = word_tokenize(quote, language=language)
    tags = nltk.pos_tag(words)
    tree = nltk.ne_chunk(tags, binary=True)
    return set(" ".join(i[0] for i in t) for t in tree if hasattr(t, "label") and t.label() == "NE")
```

```python


def spell_correct_addition(bad_word, dictionary):

    a = {}
    
    for w in dictionary:
        b = bad_word 
        r = 0
        if len(b) >= len(w): #then additions won't get you there
            r = math.inf
            a.update({w: r})
            continue
        for j, e in enumerate(w):
            if j < len(b):
                if b[j] != e:
                    b = b[:j] + e + b[j:]
                    r += 1
                    if b == w:
                        break
                    if len(b) >= len(w):
                        r = math.inf
                        break    
            else: 
                b = b[:j] + e
                r += 1
                if b == w:
                    break
                if len(b) >= len(w):
                    r = math.inf
                    break    
        a.update({w: r})
    
    return a


def spell_correct_subtraction(bad_word, dictionary):

    a = {}

    for w in dictionary:
        b = bad_word 
        r = 0
        if len(b) <= len(w): #then subtractions won't get you there
            r = math.inf
            a.update({w: r})
            continue
        for j, e in enumerate(w):
            if j < len(b):
                if b[j] == e:
                    if j == (len(w) - 1):
                        r = r + (len(b) - len(w))
                    continue
                else:
                    for k in range(j + 1, len(b)):
                        if b[k] == e:
                            b = b[:j] + b[k:]
                            break
                        elif k == len(b) - 1:    
                            b = b[:j]
                    r = r + (k - j)
                    if b == w:
                        break
                    elif len(b) <= len(w):    
                        r = math.inf
                        break
                    elif j == len(w) - 1:
                        r = r + (len(b) - len(w))
            else:
                if b != w:
                    r = math.inf
        a.update({w: r})
    
    return a


def spell_correct_substitution(bad_word, dictionary):

    a = {}
    
    for w in dictionary:
        b = bad_word 
        r = 0
        if len(b) != len(w): #then substitutions won't get you there
            r = math.inf
            a.update({w: r})
            continue
        for j, e in enumerate(w):
            if b[j] != e:
                b = b[:j] + e + b[(j+1):]
                r += 1
                if b == w:
                    break
        a.update({w: r})

    return a


def spell_correct_rearrangement(bad_word, dictionary):

    a = {}

    for w in dictionary:
        b = bad_word 
        r = 0
        if sorted(b) != sorted(w): #then rearrangements won't work, must have the exact same letters & amounts to work
            r = math.inf
            a.update({w: r})
            continue
        for j, e in enumerate(w):
            if b[j] != e:
                for k in range(len(b)):
                    if k <= j:
                        continue
                    if b[k] == e:
                        break
                r = r + (k - j)
                b = b[:j] + b[k] + b[j:k] + b[(k+1):]
                if b == w:
                    break
        a.update({w: r})

    return a


def spell_correct_split_2(bad_word, dictionary):

    a = {}
    b = bad_word 
    
    for i in range(len(b)-1):
        bup = b[:(i+1)] + ' ' + b[(i+1):] 
        r = math.inf
        if sum([l in dictionary for l in bup.split()]) == 2:
            r = 1
        a.update({bup: r})

    return a


def spell_correct_expansion_substitution(bad_word, dictionary):

    a = {}
    
    for w in dictionary:
        b = bad_word 
        if len(b) >= len(w): #then expansion won't get you there
            r = math.inf
            a.update({w: r})
            continue
        dist = len(w) - len(b)
        r = dist
        for j, e in enumerate(w):
            if j >= len(b):
                r = math.inf
                break
            if b[j] != e:
                b = b[:(j+1)] + ' ' * dist + b[(j+1):]
                break
        if r != math.inf:        
            for j, e in enumerate(w):
                if b[j] != e:
                    b = b[:j] + e + b[(j+1):]
                    r += 1
                    if b == w:
                        break
        a.update({w: r})
    
    return a



def spell_correct(bad_word, dictionary):
    
    rear = spell_correct_rearrangement(bad_word, dictionary)
    adtn = spell_correct_addition(bad_word, dictionary)
    subt = spell_correct_subtraction(bad_word, dictionary)
    subs = spell_correct_substitution(bad_word, dictionary)
    #splt = spell_correct_split_2(bad_word, dictionary)
    #expa = spell_correct_expansion_substitution(bad_word, dictionary)
    
    return [rear, adtn, subt, subs]#, splt, expa]






#Priority if all else equal:
```

```python
#Slow version

def word_alignment(b, w):

    d = {}
    r = {}
    s = []
    
    for k in w:
        if k not in s:
            s.append(k)

    for k in s:
        p = [i for i, l in enumerate(w) if k == l]
        q = [i for i, l in enumerate(b) if k == l]

        if not len(q):
            q = [-1] * len(p)
        elif len(q) < len(p):
            q2 = [-1] * len(p)
            for u in q:
                tmp = []
                for i, t in enumerate(p):
                    if q2[i] == -1:
                        tmp.append(abs(u-t))
                    else:
                        tmp.append(math.inf)
                m = tmp.index(min(tmp))
                if q2[m] == -1:
                    q2[m] = u
            q = q2
        elif len(q) > len(p):
            p2 = [-1] * len(q)
            for u in p:
                tmp = []
                for i, t in enumerate(q):
                    if p2[i] == -1:
                        tmp.append(abs(u-t))
                    else:
                        tmp.append(math.inf)
                m = tmp.index(min(tmp))
                if p2[m] == -1:
                    p2[m] = u
            p = p2

        d.update({k: p})
        r.update({k: q})
        
    return d, r


def word_alignment_penalty(b, w):

    d, r = word_alignment(b, w)
    q1 = word_alignment_expansion(d, r)
    d, r = word_alignment(w, b)
    q2 = word_alignment_expansion(d, r)

    c = 0
    for i, e in enumerate(q2):
        if e == -1:
            if i < len(q1):
                if q1[i] != -1:
                    c += 1
            else:
                c += 1
    
    return c


def word_alignment_expansion(d, r):
    #changes are being made to dictionaries d & r so copies of the supplied arguments are made since dictionaries
    #are mutable in python.
    
    w = word_expansion(d)

    dc = d.copy()
    rc = r.copy()
        
    p = [None] * len(w)
    q = [None] * len(w)
    
    for i, k in enumerate(w):
        rc[k] = [rc[k][j] for j, e in enumerate(dc[k]) if e != -1]
        dc[k] = [e for e in dc[k] if e != -1]
        p[i] = dc[k][0] 
        q[i] = rc[k][0]
        dc[k] = dc[k][1:]
        rc[k] = rc[k][1:]
    
    return p, q


def adjust_word_alignment_expansion(q):

    qs = [e for e in q if e != -1]
    qs.sort()
    qn = [0] * len(q)

    for i in range(len(q)):
        if q[i] != -1:
            j = qs.index(q[i])
            qn[i] = j
        else:
            qn[i] = -1

    return qn


def word_alignment_score_vector(q):
    
    v = [1] * len(q)
    pp = [i for i, e in enumerate(q) if e != -1]     
    qc = [e for e in q if e != -1]
    vc = [0] * len(qc)
    for i in range(len(vc)):
        if qc[i] != i:
            vc[i] = 1
            p = [j for j, e in enumerate(qc[i:]) if e == i][0] + i
            t = qc[i:p]
            qc[i] = qc[p]
            qc[(i+1):(p+1)] = t

    for i in range(len(vc)):
        v[pp[i]] = vc[i]

    return v

def word_expansion(d):
    
    w = [None] * sum([1 for k, e in d.items() for j in e if j != -1])
    for k, e in d.items():
        for j in e:
            if j != -1:
                w[j] = k
    
    return ''.join(w)
    
    
def word_alignment_score(b, w, max_diff = 3, max_penalty = 3):
    
    if abs(len(b) - len(w)) > max_diff:
        return math.inf
    
    u = word_alignment_penalty(b, w)        
    if u  > max_penalty:
        return math.inf
    
    d, r = word_alignment(b, w)
    q = word_alignment_expansion(d, r)
    q = adjust_word_alignment_expansion(q)
    v = word_alignment_score_vector(q)    

    return sum(v) + u


def word_score(b, w, max_diff = 3, max_penalty = 3):
    #Slow version


    bb = b[::-1]
    wb = w[::-1]
    
    sf = word_alignment_score(b, w, max_diff = max_diff, max_penalty = max_penalty) #[f]orward
    sb = word_alignment_score(bb, wb, max_diff = max_diff, max_penalty = max_penalty) #[b]ackward
    
    return min(sf, sb)


def spell_correction(b, dictionary):
    
    sc = {} #[sc]ores
    for w in dictionary:
        ws = word_score(b, w) #[w]ord [s]core
        sc.update({w: ws})
        
    return sc    

```

```python
#Fast version

#The next goal is to take the possible word replacements across multiple scores, not just the minimum
#and then rank the suggested replacements based on their score AND how likely it is based on how often
#that word is used in general or if possible based on the context or other words surrounding it in the sentence
# AND based on how the penalties were awarded. Rearranging one letter is more trustworthy than having to 
# add a letter, for example. 


def word_alignment(b, w):

    d = {}
    r = {}
    s = []
    
    for k in w:
        if k not in s:
            s.append(k)

    for k in s:
        p = [i for i, l in enumerate(w) if k == l]
        q = [i for i, l in enumerate(b) if k == l]

        if not len(q):
            q = [-1] * len(p)
        elif len(q) < len(p):
            q2 = [-1] * len(p)
            for u in q:
                tmp = []
                for i, t in enumerate(p):
                    if q2[i] == -1:
                        tmp.append(abs(u-t))
                    else:
                        tmp.append(math.inf)
                m = tmp.index(min(tmp))
                if q2[m] == -1:
                    q2[m] = u
            q = q2

        r.update({k: q})
        
    return r


def word_alignment_expansion(r, w):
    #changes are being made to dictionaries d & r so copies of the supplied arguments are made since dictionaries
    #are mutable in python.
    
    rc = r.copy()
    q = [None] * len(w)
    
    for i, k in enumerate(w):
        q[i] = rc[k][0]
        rc[k] = rc[k][1:]
    
    return q


def word_alignment_penalty(b, w, r):

    q1 = word_alignment_expansion(r, w)
    r = word_alignment(w, b)
    q2 = word_alignment_expansion(r, b)

    c = 0
    for i, e in enumerate(q2):
        if e == -1:
            if i < len(q1):
                if q1[i] != -1:
                    c += 1
            else:
                c += 1
    
    return c


def adjust_word_alignment_expansion(q):

    qs = [e for e in q if e != -1]
    qs.sort()
    qn = [0] * len(q)

    for i in range(len(q)):
        if q[i] != -1:
            j = qs.index(q[i])
            qn[i] = j
        else:
            qn[i] = -1

    return qn


def word_alignment_score_vector(q):
    
    v = [1] * len(q)
    pp = [i for i, e in enumerate(q) if e != -1]     
    qc = [e for e in q if e != -1]
    vc = [0] * len(qc)
    for i in range(len(vc)):
        if qc[i] != i:
            vc[i] = 1
            p = [j for j, e in enumerate(qc[i:]) if e == i][0] + i
            t = qc[i:p]
            qc[i] = qc[p]
            qc[(i+1):(p+1)] = t

    for i in range(len(vc)):
        v[pp[i]] = vc[i]

    return v


def word_expansion(d):
    
    w = [None] * sum([1 for k, e in d.items() for j in e if j != -1])
    for k, e in d.items():
        for j in e:
            if j != -1:
                w[j] = k
    
    return ''.join(w)
    

def word_alignment_score(b, w, max_diff = 3, max_penalty = 3):
    
    if abs(len(b) - len(w)) > max_diff:
        return math.inf
    
    r = word_alignment(b, w)
    u = word_alignment_penalty(b, w, r)        
    if u > max_penalty:
        return math.inf
    
    q = word_alignment_expansion(r, w)
    q = adjust_word_alignment_expansion(q)
    v = word_alignment_score_vector(q)    

    return sum(v) + u



def word_score(b, w, max_diff = 3, max_penalty = 3):
    #Fast version

    bb = b[::-1]
    wb = w[::-1]
    
    sf = word_alignment_score(b, w, max_diff = max_diff, max_penalty = max_penalty) #[f]orward
    sb = word_alignment_score(bb, wb, max_diff = max_diff, max_penalty = max_penalty) #[b]ackward
    
    return min(sf, sb)


def spell_correction(b, dictionary):
    
    sc = {} #[sc]ores
    for w in dictionary:
        ws = word_score(b, w) #[w]ord [s]core
        sc.update({w: ws})
        
    return sc    


```
